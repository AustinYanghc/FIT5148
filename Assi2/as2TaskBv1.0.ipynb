This document describes a data model that uses embedded documents to describe relationships between connected data. Consider the TaskB that maps climatedata and multiple firedata relationships. I chose embedding over referencing because if we need to view many data entities in context of another. 

```python
import pymongo
from pprint import pprint
import csv
import datetime
from pymongo import MongoClient
import pandas as pd
import csv
import json
```

```python
client = MongoClient ()
db = client.fit5148_db
```

```python
climate = db.climate
fire = db.fire
climatefire = db.climatefire
```

```python
results = climate.aggregate([
{"$unwind": "$Date"},
{
"$lookup":
    {
    "from": "fire",
    "localField": "Date",
    "foreignField": "Date",
    "as": "climate_fire"
    }    
},
    {"$out":"climatefire"}
])
```

```python
for i in results:
    pprint(i)
```

TaskB.6
.sort(), in pymongo, takes key and direction as parameters.

So if you want to sort by, let's say, id then you should .sort("_id", 1)

```python
results = climatefire.find({}, {"climate_fire.Surface Temperature (Celcius)":1,"climate_fire._id":1}).sort({"climate_fire.Surface Temperature (Celcius):1",pymongo.DESCENDING}).limit(10)
for doc in results:
    pprint(doc)
```

TaskB.7

```python
results = climatefire.aggregate([
    {"$project" : { "Date": "$climate_fire.Date", "_id" : "$_id"}},
    {"$unwind" : "$Date"},
    {"$group":{ "_id": {"date": "$Date"}, "count":{"$sum":1}}}
])
for doc in results:
    pprint(doc)
```

TaskB.8

```python
results = climatefire.aggregate([
    {"$project" : { "Date": "$climate_fire.Date", "Surface Temperature": "$climate_fire.Surface Temperature (Celcius)"}},
    {"$unwind" : "$Date"},
    {"$unwind" : "$Surface Temperature"},
    {"$group":{ "_id": {"date": "$Date"},"avgSurfaceTem":{"$avg": "$Surface Temperature"}}}
])
for doc in results:
    pprint(doc)
```
